```python
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve
from sklearn.metrics import r2_score, make_scorer
from imblearn.over_sampling import SMOTE
import collections
```

```python
df = pd.read_csv("datasets/ml_ready_set_unbalanced.csv")
```

```python
scaler = StandardScaler()
def scale_data(X_train, X_test):
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.fit_transform(X_test)
    return X_train_scaled, X_test_scaled
```

```python
target = df["case_status_encoded"]
df_model = df.drop("case_status_encoded", axis=1)
X = df_model
y = target
```

```python
train_sizes, train_scores, valid_scores = learning_curve(LogisticRegression(), X, y, 
                                                         train_sizes=range(200, 1000, 10), 
                                                         cv=5,
                                                         scoring=make_scorer(r2_score))
```

```python
plt.style.use('fivethirtyeight')
```

```python
ax = pd.Series(train_scores.mean(axis=1)).to_frame().set_index(train_sizes)\
                .rename(columns={0: 'Train'}).plot.line()
pd.Series(valid_scores.mean(axis=1)).to_frame().set_index(train_sizes).rename(columns={0: 'Test'})\
                .plot.line(ax=ax)
```

learning curve shows high varience, not 100 percentage confident code is correct

The majority of wgae type is yearly so the other types will be converted to yearly

```python
df.loc[df.WAGE_UNIT_OF_PAY == "Week"]
```

```python
for i, row in df.iterrows():
    if row["WAGE_UNIT_OF_PAY"] == "Week":
        new_wage = df.at[i, "WAGE_RATE_OF_PAY_FROM"] * 52
        df.at[i,"WAGE_RATE_OF_PAY_FROM"] = new_wage
        df.at[i,"WAGE_UNIT_OF_PAY"] = "Year"
        
    elif row["WAGE_UNIT_OF_PAY"] == "Hour":
        new_wage = (df.at[i, "WAGE_RATE_OF_PAY_FROM"] * 40) * 52
        df.at[i,"WAGE_RATE_OF_PAY_FROM"] = new_wage
        df.at[i,"WAGE_UNIT_OF_PAY"] = "Year"
        
    elif row["WAGE_UNIT_OF_PAY"] == "Bi-Weekly":
        new_wage = df.at[i, "WAGE_RATE_OF_PAY_FROM"] * 26
        df.at[i,"WAGE_RATE_OF_PAY_FROM"] = new_wage
        df.at[i,"WAGE_UNIT_OF_PAY"] = "Year"
        
    elif row["WAGE_UNIT_OF_PAY"] == "Month":
        new_wage = df.at[i, "WAGE_RATE_OF_PAY_FROM"] * 12
        df.at[i,"WAGE_RATE_OF_PAY_FROM"] = new_wage
        df.at[i,"WAGE_UNIT_OF_PAY"] = "Year"
```
